## Menentukan Pertanyaan Bisnis

**Informasi data set**

Data set berasal dari data polusi udara dari 12 Stasiun Pengamatan di China yaitu Aotizhongxin, Changping, Dingling, Dongsi, Guanyuan, Gucheng, Huairou, Nongzhanguan, Shunyi, Tiantan, Wanliu, dan Wanshouxigong. Periode pengamatan yaitu dari March 1, 2013 hingga February 28, 2017.

Data set terdiri dari 18 feature yaitu :
* No: Nomor urut data
* year: Tahun pengamatan
* month: Bulan pengamatan
* day: Hari pengamatan
* hour: Jam pengamatan
* PM2.5 (Particulate Matter 2.5): Konsentrasi partikel kecil dengan diameter kurang dari 2.5 mikrometer dalam udara (satuan: mikrogram per meter kubik)
* PM10 (Particulate Matter 10): Konsentrasi partikel dengan diameter kurang dari 10 mikrometer dalam udara (satuan: mikrogram per meter kubik).
* SO2 (Sulfur Dioxide): Konsentrasi gas sulfur dioksida dalam udara (satuan: mikrogram per meter kubik)
* NO2 (Nitrogen Dioxide): Konsentrasi gas nitrogen dioksida dalam udara (satuan: mikrogram per meter kubik).
* CO (Carbon Monoxide): Konsentrasi gas karbon monoksida dalam udara (satuan: miligram per meter kubik).
* O3 (Ozone): Konsentrasi gas ozon dalam udara (satuan: mikrogram per meter kubik).
* TEMP (Temperature): Suhu udara (satuan: derajat Celsius).
* PRES (Pressure): Tekanan atmosfer (satuan: hektopascal atau milibar).
* DEWP (Dew Point): Titik embun udara (satuan: derajat Celsius).
* RAIN (Rainfall): Curah hujan (satuan: millimeter)
* wd (Wind Direction): Arah angin (satuan: derajat).
* WSPM (Wind Speed): Kecepatan angin (satuan: meter per detik).
* station: Nama stasiun pengamatan polusi udara.


Dari data set ini pemerintah perlu mengevaluasi kebijakan terkait lingkungan untuk meningkatkan kulitas hidup masyarakat. Selain itu, dengan memahami kondisi polusi udara pemerintah dapat melakukan inovasi dan terobosan baru yang sesuai berdasarkan data yang ada.

### Cleaning Data

```python
dataframe = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12]
station = ["Aotizhongxin", "Changping", "Dingling", "Dongsi", "Guanyuan", "Gucheng", "Huairou", "Nongzhanguan", "Shunyi", "Tiantan", "Wanliu", "Wanshouxigong"]
type_num = ["PM2.5", "PM10", "SO2", "NO2", "CO", "O3", "TEMP", "PRES", "DEWP", "RAIN", "WSPM"]
type_obj = ["wd"]
```

#### Mengisi missing value

Untuk mengisi missing value, data dengan tipe numerik digunakan interpolasi linear sedangkan untuk data objek digunakan modus.

```python
def missing_cleaning(data):
  for i in type_num :
    data[i].interpolate(method='linear', limit_direction='forward', inplace=True)

  for j in type_obj:
    data[j].fillna(value=data[j].mode()[0], inplace=True)
```

```python
for i in dataframe:
  missing_cleaning(i)
```

```python
df1.isna().sum()
```

**Interpretasi :**
Sudah tidak ada missing value.

#### Mengatasi Outlier

Untuk mengatasi outlier dinggunakan boundary value untuk mengganti nilai outlier atas dan bawah.

```python
def outlier_cleaning(data):
  for i in type_num :
    Q1 = (data[i]).quantile(0.25)
    Q3 = (data[i]).quantile(0.75)
    IQR = Q3 - Q1

    maximum = Q3 + (1.5*IQR)
    minimum = Q1 - (1.5*IQR)

    kondisi_lower_than = data[i] < minimum
    kondisi_more_than = data[i] > maximum

    data[i] = data[i].mask(cond=kondisi_more_than, other=maximum)
    data[i] = data[i].mask(cond=kondisi_lower_than, other=minimum)
```

```python
outlier_cleaning(df1)
```

```python
df1.describe()
```

**Interpretasi :**
Niali maksimum dan minimum sudah wajar yang mengindikasikan sudah tidak ada outlier.

## Exploratory Data Analysis (EDA)

Petanyaan analisis yang dibuat keduanya melibatkan data kuantitas yang hanya berfokus pada "PM2.5", "PM10", "SO2", "NO2", "CO", dan "O3". Selanjutnya untuk EDA hanya difokuskan pada 6 komponen tersebut

```python
def group_mean (data):
  return data.groupby(by="year").agg({
      "PM2.5": ["sum","mean"],
      "PM10": ["sum","mean"],
      "SO2": ["sum","mean"],
      "NO2": ["sum","mean"],
      "CO": ["sum","mean"],
      "O3": ["sum","mean"],
  })
```

```python
def group_minmax (data):
  return data.groupby(by="year").agg({
      "PM2.5": ["min","max"],
      "PM10": ["min","max"],
      "SO2": ["min","max"],
      "NO2": ["min","max"],
      "CO": ["min","max"],
      "O3": ["min","max"],
  })
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana jumlah polutan di setiap stasiun pengamatan tiap tahunnya selama periode pengamatan?

Terlebih dahulu dibuat data frame yang dikelompokkan berdasarkan tahun lalu dicari total polutan pada tahun itu. Selanjutnya, data frame tersebut diubah posisinya dimana kolom polutan diubah menjadi baris dengan bantuan pd.melt(). Hasil visualisasi sebanyak 12.

```python
for (data, loc) in zip(dataframe, station):
  # Group data berdasarkan tahun dan kemudian menghitung total masing-masing polutan
  grouped_data = data.groupby('year').sum()

  # Reset index untuk membuat kolom 'year' menjadi kolom biasa
  grouped_data.reset_index(inplace=True)

  # Melt dataframe untuk membuat kolom polutan menjadi baris
  melted_data = pd.melt(grouped_data, id_vars=['year'], value_vars=['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3'],
                        var_name='pollutant', value_name='sum_value')

  # Plot menggunakan seaborn
  plt.figure(figsize=(10, 6))
  sns.barplot(data=melted_data, x="year", y="sum_value", hue="pollutant")
  plt.xlabel('Tahun')
  plt.ylabel('Total konsentrasi (ug/m^3)')
  plt.title('Total Polutan per tahun di ' + loc)
  plt.legend(loc='upper right')
  plt.show()
```

### Pertanyaan 2: Bagaimana tren konsentrasi polutan PM2.5 dan PM10 selama satu bulan dan satu minggu terakhir pada periode pengamatan di setiap stasiun pengamatan?

Untuk mengetahui tren pada periode tertentu diperlukan data start_date dan end_date.

Untuk periode satu bulan terakhir digunakan
* start_date = "2017-02-01 02:00:00"
* end_date = "2017-02-28 02:00:00"  

Untuk periode satu minggu terakhir digunakan
* start_date = "2017-02-21 02:00:00"
* end_date = "2017-02-28 02:00:00"

Apabila menghendaki rentan periode tertentu bisa diganti variabel start_date dan end_date dengan format di atas.

```python
# Tren PM2.5 Bulan Februari 2017
start_date = "2017-02-01 02:00:00"
end_date = "2017-02-28 02:00:00"

colors = ['red', 'blue', 'green', 'orange', 'purple', 'cyan', 'magenta', 'yellow', 'brown', 'pink', 'teal', 'lime']

plt.figure(figsize=(12, 5))
for i, (data, loc) in enumerate(zip(dataframe, station)):
    main_df = data[(data['datetime'] >= str(start_date)) & (data['datetime'] <= str(end_date))]
    plt.plot(main_df['datetime'], main_df['PM2.5'], label=loc, color=colors[i % len(colors)])

plt.title('Tren Konsentrasi PM2.5 Bulan Februari 2017', size=20)
plt.xlabel('Date',size=13)
plt.ylabel('Konsentrasi PM2.5 (ug/m^3)',size=13)
plt.legend()
plt.show()

# Tren PM10 Bulan Februari 2017
plt.figure(figsize=(12, 5))
for i, (data, loc) in enumerate(zip(dataframe, station)):
    main_df = data[(data['datetime'] >= str(start_date)) & (data['datetime'] <= str(end_date))]
    plt.plot(main_df['datetime'], main_df['PM10'], label=loc, color=colors[i % len(colors)])

plt.title('\nTren Konsentrasi PM10 Bulan Februari 2017', size=20)
plt.xlabel('Date',size=13)
plt.ylabel('Konsentrasi PM10 (ug/m^3)',size=13)
plt.legend()
plt.show()
```

```python
# Tren PM2.5 Minggu Terakhir Februari 2017
start_date = "2017-02-21 02:00:00"
end_date = "2017-02-28 02:00:00"

colors = ['red', 'blue', 'green', 'orange', 'purple', 'cyan', 'magenta', 'yellow', 'brown', 'pink', 'teal', 'lime']

plt.figure(figsize=(12, 5))
for i, (data, loc) in enumerate(zip(dataframe, station)):
    main_df = data[(data['datetime'] >= str(start_date)) & (data['datetime'] <= str(end_date))]
    plt.plot(main_df['datetime'], main_df['PM2.5'], label=loc, color=colors[i % len(colors)])

plt.title('Tren Konsentrasi PM2.5 Minggu Terakhir Februari 2017', size=20)
plt.xlabel('Date',size=13)
plt.ylabel('Konsentrasi PM2.5 (ug/m^3)',size=13)
plt.legend()
plt.show()

# Tren PM10 Minggu Terakhir Februari 2017
plt.figure(figsize=(12, 5))
for i, (data, loc) in enumerate(zip(dataframe, station)):
    main_df = data[(data['datetime'] >= str(start_date)) & (data['datetime'] <= str(end_date))]
    plt.plot(main_df['datetime'], main_df['PM10'], label=loc, color=colors[i % len(colors)])

plt.title('\nTren Konsentrasi PM10 Minggu Terakhir Februari 2017', size=20)
plt.xlabel('Date',size=13)
plt.ylabel('Konsentrasi PM10 (ug/m^3)',size=13)
plt.legend()
plt.show()
```
